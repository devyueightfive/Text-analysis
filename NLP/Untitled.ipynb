```python
import numpy as np
import  matplotlib.pyplot as plt
```

```python
def fx(w, t):
    return 0.129*np.sin(w*t)

def fy(w,t,fi):
    return -0.099*np.cos(w*t+fi)
```

```python
T = 2
ts = np.linspace(0, T, 1000)

w = 2*np.pi/T
fi = np.pi/3

X = fx(w, ts)
Y = fy(w, ts, fi)

plt.figure(figsize = (8,8))
plt.grid(True, which = 'both')
plt.plot(X, Y)
plt.show()
```
